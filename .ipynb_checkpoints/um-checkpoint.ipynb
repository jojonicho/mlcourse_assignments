```python
import pandas as pd

def IMPOR(filename):
    df = pd.read_csv(filename, names = ['nama','jenis','asal','link'])
    return df
def CARINAMA(nama):
    q = '(nama == "'+nama+'")'
    return df.query(q).to_csv(header=False, index=False)
def CARITIPE(tipe):
    q = '(jenis == "'+tipe+'")'
    return df.query(q).to_csv(header=False, index=False), len(df.query(q))
def CARIPROV(prov):
    q = '(asal == "'+prov+'")'
    return df.query(q).to_csv(header=False, index=False), len(df.query(q))
def TAMBAH(row):
    args = row.split(';;;')
    df.loc[len(df)] = [args[i] for i in range(4)]
    return df, args[0]
def UPDATE(row):
    args = row.split(';;;')
    q = '(nama == "'+args[0]+'")'
    index = (df.query(q).index)
    df.iloc[index,:] = [args[i] for i in range(4)]
    return df, args[0]
def HAPUS(nama):
    return df[df.nama != nama]
def STATTIPE():
    stattipe = []
    jenis = df.jenis.unique()
    for jen in jenis:
        stattipe.append((jen, len(df[df.jenis ==jen])))
    return stattipe
def STATPROV():
    statprov = []
    locs = df.asal.unique()
    for loc in locs:
        statprov.append((loc, len(df[df.asal ==loc])))
    return statprov
def EKSPOR(filename):
    df.to_csv(header=False, index=False, path_or_buf=filename)
    return ('Terekspor %d baris' %len(df))

print("#####\nBudayaKB Lite v1.0\nKalau bukan kita yang melestarikan budaya, siapa lagi?\n#####")
a = input("Masukkan perintah: ").split()
perintah = a[0]
arg = ' '.join(a[1:])
while perintah != 'KELUAR':
    if perintah == 'IMPOR':
        df = IMPOR(arg)
        print('Terimpor %d baris\n#####' % len(df))
    elif perintah == 'CARINAMA':
        print(CARINAMA(arg), end='')
        print('#####')
    elif perintah == 'CARITIPE':
        csv, macam = CARITIPE(arg)
        print(csv, end='')
        print('*Ditemukan %d %s*\n#####'%(macam, arg))
    elif perintah == 'CARIPROV':
        csv, macam = CARIPROV(arg)
        print(csv, end='')
        print('*Ditemukan %d warisan budaya*\n#####'%(macam))
    elif perintah == 'TAMBAH':
        df, name = TAMBAH(arg)
        print("%s telah ditambahkan" %name)
        print('#####')
    elif perintah == 'UPDATE':
        df, name = UPDATE(arg)
        print("%s telah diupdate\n#####" %name)
    elif perintah == 'HAPUS':
        df = HAPUS(arg)
        print('%s dihapus\n#####' %arg)
    elif perintah == 'STAT':
        print('Terdapat %d warisan budaya\n#####' %len(df))
    elif perintah == 'STATTIPE':
        print(STATTIPE())
        print('#####')
    elif perintah == 'STATPROV':
        print(STATPROV())
        print('#####')
    elif perintah == 'EKSPOR':
        print(EKSPOR(arg))
        print('#####')
    elif perintah == 'df':
        print(df)
    a = input("Masukkan perintah: ").split()
    perintah = a[0]
    arg = ' '.join(a[1:])
print('~Sampai jumpa, jangan lupa mencintai warisan budaya Indonesia!~')
```
